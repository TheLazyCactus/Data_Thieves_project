```python
import requests
from dotenv import load_dotenv
import os
import pandas as pd
```

```python
load_dotenv()
```

##### Retirement/Nursing Homes

```python
api_key = os.getenv("api_key")

keyword = 'retirement+home'
location = '34.056455207345294,-118.24834617606246'
radius = '50000'
type = 'nursing_home|health|establishment|hospital'

url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?keyword={keyword}&location={location}&radius={radius}&type={type}&key={api_key}"
```

```python
response = requests.get(url)
response.status_code
```

```python
response.content
```

```python
info = response.json()
info
```

```python
retirement_homes_df = pd.json_normalize(info['results'], sep='_')
pd.set_option('display.max_columns', None)
retirement_homes_df
```

```python
retirement_homes_df = retirement_homes_df.drop(['icon', 'icon_background_color', 'icon_mask_base_uri', 'reference', 'scope', 'geometry_location_lat', 'geometry_location_lng', 'geometry_viewport_northeast_lat', 'geometry_viewport_northeast_lng', 'geometry_viewport_southwest_lat', 'geometry_viewport_southwest_lng', 'opening_hours_open_now', 'plus_code_global_code'], axis=1)
retirement_homes_df.head()
```

##### Schools

```python
keyword_2 = 'school'
location_2 = '34.056455207345294,-118.24834617606246'
radius_2 = '50000'
type_2 = 'school|university'

url_2 = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?keyword={keyword_2}&location={location_2}&radius={radius_2}&type={type_2}&key={api_key}"
```
