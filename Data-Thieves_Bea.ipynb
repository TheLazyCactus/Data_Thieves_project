```python
load_dotenv()
```

##### Retirement/Nursing Homes

```python
api_key = os.getenv("api_key")

keyword = 'retirement+home'
location = '34.056455207345294,-118.24834617606246'
radius = '50000'
type = 'nursing_home|health|establishment|hospital'

url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?keyword={keyword}&location={location}&radius={radius}&type={type}&key={api_key}"
```

```python
response = requests.get(url)
response.status_code
```

```python
response.content
```

```python
all_results = []
while url:
    response = requests.get(url)
    if response.status_code != 200:
        print("Error:", response.status_code)
        break 
    info = response.json()
    results = info.get('results', [])
    all_results.extend(results)

    next_page_token = info.get('next_page_token')
    if next_page_token:
        time.sleep(2)
        url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?pagetoken={next_page_token}&key={api_key}"
    else:
        break
```

```python
retirement_homes_df = pd.json_normalize(all_results, sep='_')
pd.set_option('display.max_columns', None)
retirement_homes_df
```

```python
retirement_homes_df = retirement_homes_df.drop(['icon', 'business_status', 'photos', 'icon_background_color', 'user_ratings_total', 'icon_mask_base_uri', 
                                                'reference', 'scope', 'geometry_location_lat', 'geometry_location_lng', 
                                                'geometry_viewport_northeast_lat', 'geometry_viewport_northeast_lng', 
                                                'geometry_viewport_southwest_lat', 'geometry_viewport_southwest_lng', 
                                                'opening_hours_open_now', 'plus_code_global_code'], axis=1
)
retirement_homes_df.head()
```

```python
place_ids = retirement_homes_df['place_id']

def get_phone_number(place_id):
    details_url = f"https://maps.googleapis.com/maps/api/place/details/json?place_id={place_id}&fields=formatted_phone_number&key={api_key}"
    details_response = requests.get(details_url)
    details_info = details_response.json()

    return details_info['result'].get('formatted_phone_number', None)

retirement_homes_df['phone_number'] = [get_phone_number(place_id) for place_id in place_ids]

retirement_homes_df.head()
```

```python
def get_schedule(place_id):
    details_url = f"https://maps.googleapis.com/maps/api/place/details/json?place_id={place_id}&fields=opening_hours&key={api_key}"
    details_response = requests.get(details_url)
    details_info = details_response.json()

    if 'result' in details_info and 'opening_hours' in details_info['result']:
        weekday_text = details_info['result']['opening_hours'].get('weekday_text')
        
        if weekday_text:
            first_day = weekday_text[0]  # Extract Monday's hours
            time_range = first_day.split(': ')[1].strip()  # e.g., "9:00 AM  5:00 PM"
            
            # Check if time_range contains an en-dash and handle accordingly
            if '' in time_range:
                opening_time, closing_time = time_range.split('')
                return opening_time.strip(), closing_time.strip()
            else:
                # Return 'N/A' if the time range is not in the expected format
                return 'N/A', 'N/A'

    return 'N/A', 'N/A'

retirement_homes_df[['opening_time', 'closing_time']] = pd.DataFrame([get_schedule(place_id) for place_id in place_ids])

retirement_homes_df.head()
```
