```python
import requests
from dotenv import load_dotenv
import os
import pandas as pd
```

```python
load_dotenv()
```

##### Retirement/Nursing Homes

```python
api_key = os.getenv("api_key")

keyword = 'retirement+home'
location = '34.056455207345294,-118.24834617606246'
radius = '50000'
type = 'nursing_home|health|establishment|hospital'

url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?keyword={keyword}&location={location}&radius={radius}&type={type}&key={api_key}"
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
response.content
```

```python
info = response.json()
```

```python
info
```

```python
df = pd.json_normalize(info, sep='_')
df.head()
```

```python
df_updated = pd.json_normalize(df['results'].explode(), sep='_')
df_updated
```
